```python
url = ("https://www.flipkart.com/search?q=lenovo%20laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import csv
```

```python
requests.get(url)
```

```python
data = requests.get(url)
BeautifulSoup(data.content)
```

```python
soup = BeautifulSoup(data.content)
```

```python
soup.find_all("div",class_="_4rR01T")[0].text
```

```python

Lapi_name = []
Price = []
MRP_Price = []
Discound = []
Rating = []
```

```python
for i in soup.findAll("a",href=True,attrs={"class":"_1fQZEK"}):
    lapi_name = soup.find("div",attrs={"class":"_4rR01T"})
    price = soup.find("div",attrs={"class":"_30jeq3 _1_WHN1"})
    mrp_Price = soup.find("div",attrs={"class":"_3I9_wc _27UcVY"})
    discound = soup.find("div",attrs={"class":"_3Ay6Sb"})
    rating = soup.find("div",attrs={"class":"_3LWZlK"})
    
    Lapi_name.append(lapi_name.text)
    Price.append(price.text)
    MRP_Price.append(mrp_Price.text)
    Discound.append(discound.text)
    Rating.append(rating.text)
    
```

```python
import pandas as pd

pd.DataFrame({"lapi_name":Lapi_name,"price":Price,"mrp_Price":MRP_Price,"rating":Rating})
```

```python

df = pd.DataFrame({"lapi_name":Lapi_name,"price":Price,"mrp_Price":MRP_Price,"rating":Rating})

df.to_csv("Filp_Prod_List")
```
