```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
urls = ("https://www.flipkart.com/laptops/lenovo~brand/pr?sid=6bo,b5g")
```

```python
# Urls Requests/Responce Check

requests.get(urls)
```

```python
data = requests.get(urls)
soup = BeautifulSoup(data.content)

soup
```

```python
# fetch single product of name

heading = soup.find_all("div",class_="_4rR01T")[0].text
heading
```

```python
# Fetch multiple product details
```

```python
Lapi_name = []
Price = []
MRP_Price = []
Discound = []
Ratings = []
```

```python
for a in soup.findAll("a",href=True,attrs={"class":"_1fQZEK"}):
    lapi_name = a.find("div",attrs={"class":"_4rR01T"})
    price = a.find("div",attrs={"class":"_30jeq3 _1_WHN1"})
    mrp_price = a.find("div",attrs={"class":"_3I9_wc _27UcVY"})
    discound = a.find("div",attrs={"class":"_3Ay6Sb"})
    ratings = a.find("div",attrs={"class":"_3LWZlK"})
    
    Lapi_name.append(lapi_name.text)
    Price.append(price.text)
    MRP_Price.append(mrp_price.text)
    Discound.append(discound.text)
    Ratings.append(ratings.text)
    
```

```python
import pandas as pd
pd.DataFrame({"lapi_name":Lapi_name,"price":Price,"mrp_price":MRP_Price,"discound":Discound,"ratings":Ratings})
```

```python
#Export Data in CSV file Formate
```

```python
df = pd.DataFrame({"lapi_name":Lapi_name,"price":Price,"mrp_price":MRP_Price,"discound":Discound,"ratings":Ratings})

df.to_excel("flip_prod_list.xlsx")
```
