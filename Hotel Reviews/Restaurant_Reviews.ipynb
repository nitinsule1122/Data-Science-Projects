## Remove Special Characters 

````````python
Q = "Wow... Loved this place.!!!```````"

from re import sub
sub("[^a-zA-Z0-9 ]","",Q)
````````

````````python
q = []

for i in A.Review:
    w = sub("[^a-zA-Z0-9 ]","",i)
    w = w.upper()
    q.append(w)
````````

````````python
q[0:4]
````````

## Data Convert String to Numbers

````````python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
# data = cv.fit_transform(q).()
X = cv.fit_transform(q).toarray()
````````

````````python
Y = A.Liked
````````

## Split Data into train & test

````````python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size=0.2,random_state=21)
````````

## Using DecisionTreeClassifier

````````python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state=21,max_depth=3)
model = dtc.fit(xtrain,ytrain)
pred = model.predict(xtest)

from sklearn.metrics import accuracy_score
accuracy_score(ytest,pred)
````````

````````python
from sklearn.metrics import confusion_matrix
confusion_matrix(ytest,pred)
````````

## Do review classification using user input

````````python
rev = input("Enter Your Review: ")
inp_text = sub("[^a-zA-Z0-9 ]","",rev.lower())
X1 = cv.transform([inp_text]).toarray()
e = model.predict(X1)[0]
if(e == 0):
    print("Disliked")
else:
    print("Liked")
````````
